# Redes Neurais Artificiais

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./credit_data.csv')
df.head()
```

```python
df = df.dropna()
```

```python
X_credit = df[['income', 'age', 'loan']]
y_credit = df['default']
```

Normalizacao dos dados

```python
standard_scaler = StandardScaler()
X_credit = standard_scaler.fit_transform(X_credit)
X_credit
```

Separacao dos dados de teste e de treino

```python
X_credit_treinamento, X_credit_teste, y_credit_treinamento, y_credit_teste = train_test_split(X_credit, y_credit, test_size=0.25, random_state=0)
```

Treinamento do Modelo

```python
from sklearn.neural_network import MLPClassifier
```

```python
rede_neural = MLPClassifier(
  verbose=True, 
  max_iter=1000, 
  tol=0.0000100,
  hidden_layer_sizes=(20,100)
)
rede_neural.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = rede_neural.predict(X_credit_teste)
previsoes
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
acc = accuracy_score(y_credit_teste, previsoes)
acc
```
